```python
account_holder_name = "Jane Doe"
account_balance = 0.0

def check_balance():
    """Function to display the current balance."""
    print(f"Account Holder: {account_holder_name}")
    print(f"Current Balance: ${account_balance:.2f}")
    print("-" * 30)

def deposit(amount):
    """Function to simulate a deposit operation."""
    global account_balance
    if amount > 0:
        account_balance += amount
        print(f"Action: Deposited ${amount:.2f}")
        print(f"New Balance: ${account_balance:.2f}")
    else:
        print("Action Failed: Deposit amount must be positive.")
    print("-" * 30)

def withdraw(amount):
    """Function to simulate a withdrawal operation with conditional checks."""
    global account_balance
    if amount <= 0:
        print("Action Failed: Withdrawal amount must be positive.")
    elif account_balance >= amount:
        account_balance -= amount
        print(f"Action: Withdrew ${amount:.2f}")
        print(f"New Balance: ${account_balance:.2f}")
    else:
        print(f"Action Failed: Insufficient balance to withdraw ${amount:.2f}!")
        print(f"Current Balance: ${account_balance:.2f}")
    print("-" * 30)

print("=== BANK ACCOUNT MANAGER START ===")
check_balance()
deposit(1000.0)
withdraw(400.0)
withdraw(800.0)
check_balance()
```

```python
weekly_weather_data = [
    {"day": "Monday", "temperature": 32, "humidity": 55},
    {"day": "Tuesday", "temperature": 22, "humidity": 85},
    {"day": "Wednesday", "temperature": 18, "humidity": 92},
    {"day": "Thursday", "temperature": 28, "humidity": 65},
    {"day": "Friday", "temperature": 35, "humidity": 45}
]

def determine_weather_condition(temp, humidity):
    """Implement conditional statements to determine weather conditions."""
    if temp >= 30 and humidity < 60:
        return "Sunny"
    elif temp < 20 and humidity >= 90:
        return "Rainy"
    elif temp >= 20 and humidity >= 80:
        return "Humid"
    else:
        return "Mild"

def analyze_weather(data):
    """Function to calculate average temperature and humidity, and print results."""
    total_temp = 0
    total_humidity = 0
    num_days = len(data)

    print("=== DAILY WEATHER ANALYSIS ===")
    for day_data in data:
        temp = day_data["temperature"]
        humidity = day_data["humidity"]
        day = day_data["day"]


        condition = determine_weather_condition(temp, humidity)

        total_temp += temp
        total_humidity += humidity


        print(f"{day}: {temp}°C, Humidity: {humidity}%. Condition: {condition}")

    print("-" * 30)


    avg_temp = total_temp / num_days
    avg_humidity = total_humidity / num_days

    print("=== OVERALL AVERAGES ===")
    print(f"Average Temperature: {avg_temp:.2f}°C")
    print(f"Average Humidity: {avg_humidity:.2f}%")

analyze_weather(weekly_weather_data)
```
